# Como gravar TPCs no github

-----


1. Comear por fazer Signup e login na conta github (www.github.com)
2. Criar um repositorio em "New" tal como mostra a figura (ou semehante):



3. Ao criar o Repositorio **no esquecer marcar "Initialize this repository with a README"**



4. Carregar em **Create**. Pode demorar algum tempo (uns minutos) a ficar disponivel. De seguida carregar em **upload files**



e adicionar o(s) TPC(s)


5. Finalmente devem enviar-me um email com o URL do repositorio (onde diz nunobbras deve dizer o vosso user do github):
